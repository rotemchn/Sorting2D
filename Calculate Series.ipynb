```python
df = pd.read_csv('random_students_trials.csv')
df
```

```python
series_idx, trial, positivity, length = [], [], [], []
idx = 1
last_row = {}
for index, row in df.iterrows():
    add_series = 0
    if index > 0:
        if last_row['trial'] != row['trial']:
            series_idx.append(idx)
            trial.append(last_row['trial'])
            positivity.append(last_row['positivity'])
            length.append(last_row['trend'])
            idx=1
            #print("appended: trial:{} idx:{} len:{}".format(last_row['trial'], idx, last_row['trend']))
        elif row['positivity'] != last_row['positivity']:
            series_idx.append(idx)
            trial.append(last_row['trial'])
            positivity.append(last_row['positivity'])
            length.append(last_row['trend'])
            idx+=1
    last_row = row.copy()
```

```python
trend_df = pd.DataFrame.from_dict({'trial': trial, 'series_idx': series_idx, 'positivity': positivity, 'length':length})
trend_df
```
